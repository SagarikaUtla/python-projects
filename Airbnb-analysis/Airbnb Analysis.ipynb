```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### **DATA IMPORTING**

```python
data = pd.read_csv("airbnb dataset.csv")
data.head(20)
```

### **DATA SUMMARY**

```python
data.info()
```

```python
data.shape
```

### **DATA CLEANING**

```python
data.isnull().sum()

# dropping the null values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.duplicated().sum()
# dropping the duplicates
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# changing the data types
data.dtypes

data['id'] =data['id'].astype(object)

data['host_id'] =data['host_id'].astype(object)
data.dtypes
```

### **DATA ANALYSIS**

```python
df= data[data['price']  < 1500]
sns.boxplot(data=df, x='price')
```

```python
# Price Distribution

plt.figure(figsize=(7,4))
sns.histplot(data=df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.histplot(data=df, x='availability_365', bins=100)
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
#### avg price per bed

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
df.columns
```

```python
# price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type', palette={'Private room': 'pink',
                                                                                   'Entire home/apt': 'yellow',
                                                                                   'Hotel room': 'purple',
                                                                                   'Shared room': 'skyblue'})
                                                                                
```

```python
# distribution of room type

room_counts = df['room_type'].value_counts()


colors = ['#AEEA94', '#B771E5', '#4635B1', '#16C47F']


plt.figure(figsize=(5,5))
plt.pie(room_counts, labels=room_counts.index, colors=colors, autopct='%1.1f%%', startangle=15)
plt.title('Distribution of Room Types')
plt.show()
```

```python
# total availability by room type

availability_by_room = df.groupby('room_type')['availability_365'].sum()
colors = ['#8ecae6', '#219ebc', '#ffb703', '#fb8500'] 
plt.figure(figsize=(4,4))
plt.pie(availability_by_room, labels=availability_by_room.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Availability by Room Type')
plt.show()
```

```python
# Top 10 hosts by number of listings
top_hosts = df['host_name'].value_counts().head(10)

plt.figure(figsize=(6,4))
sns.barplot(x=top_hosts.values, y=top_hosts.index, hue=top_hosts.index,  palette="coolwarm", legend=False)
plt.xlabel("Number of Listings")
plt.ylabel("Host Name")
plt.title("Top 10 Hosts by Number of Listings")
plt.tight_layout()
plt.show()
```

```python
# average rating per room type

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
avg_rating = df.groupby('room_type')['rating'].mean().sort_values()

plt.figure(figsize=(5,4))
sns.barplot(x=avg_rating.index, y=avg_rating.values, hue=avg_rating.index, palette='pastel', legend=False)
plt.title('Average Rating per Room Type')
plt.ylabel('Average Rating')
plt.xlabel('Room Type')
plt.ylim(0, 5)
plt.show()
```

```python
# pairplot of price, number of reviews,  availability_365
pairplot_data = df[['price', 'number_of_reviews', 'availability_365', 'room_type']].dropna()

sns.pairplot(pairplot_data, vars=['price', 'number_of_reviews', 'availability_365'], hue='room_type', palette='Set2')
plt.suptitle('Pairplot of Price, Number of Reviews, and Availability by Room Type', y=1.02)
plt.show()
```

```python
# colour-code by room type
map_df = df[['latitude', 'longitude', 'room_type', 'price']].dropna()
map_df = map_df[map_df['price'] < 500]  # Optional: remove extreme outliers

room_type_colors = {
    'Entire home/apt': '#66c2a5',
    'Private room': '#fc8d62',
    'Shared room': '#8da0cb',
    'Hotel room': '#e78ac3'
}
colors = map_df['room_type'].map(room_type_colors)
plt.figure(figsize=(8,4))
plt.scatter(map_df['longitude'], map_df['latitude'], c=colors, s=10, alpha=0.5)
plt.title('Airbnb Listings Color-Coded by Room Type')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
